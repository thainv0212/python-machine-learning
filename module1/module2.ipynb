## Supervised Learning: Datasets

```python
%matplotlib notebook
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
plt.figure()
plt.title('Sample regression problem with one input variable')
X_R1, y_R1 = make_regression(n_samples=100, n_features=1,n_informative=1, 
                             bias=150, noise=30, random_state=0)
plt.scatter(X_R1, y_R1, marker='o', s=50)
plt.show()
```

```python
from sklearn.datasets import make_classification
```

```python
plt.figure()
X_R2, y_R2 = make_classification(n_samples=100, n_features=2, n_informative=2,
                                n_redundant=0, n_clusters_per_class=1, flip_y=0.1, class_sep=0.5, random_state=0)
plt.scatter(X_R2[:,0], X_R2[:,1], c=y_R2, marker='o', s=50)
plt.show()
# len(X_R2), len(y_R2)
```
