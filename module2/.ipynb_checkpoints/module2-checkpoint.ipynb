### Linear Regression: Ridge, Lasso, and Polynomial Regression

```python
%matplotlib notebook
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import make_classification
scaler = MinMaxScaler()
```

```python
plt.figure()
X_R2, y_R2 = make_classification(n_samples=100, n_features=2, n_informative=2,
                                n_redundant=0, n_clusters_per_class=1, flip_y=0.1, class_sep=0.5, random_state=0)
plt.scatter(X_R2[:,0], X_R2[:,1], c=y_R2, marker='o', s=50)
plt.show()
```

```python
scaler.fit(X_R2)
X_train_scaled = scaler.transform(X_R2)
plt.figure()
plt.scatter(X_train_scaled[:,0], X_train_scaled[:,1], c=y_R2, marker='o', s=50)
plt.show()
```

### Logistic Regression

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LogisticRegression
from adspy_shared_utilities import plot_class_regions_for_classifier_subplot
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_table('fruit_data_with_colors.txt')
df.head()
```

```python
Y_apple = (df['fruit_label'] == 1).astype(int).values
X_data = df[['width', 'height']].values
```

```python
fig, subaxes = plt.subplots(1,1, figsize=(7,5))
X_train, X_test, y_train, y_test = train_test_split(X_data, Y_apple, random_state=0)
clf = LogisticRegression(C=1000).fit(X_train, y_train)
plot_class_regions_for_classifier_subplot(clf, X_train, y_train, None,
                                         None, 'Logistic Regression \
                                         for binary classification \nFruit dataset: Apple vs others',
                                         subaxes)
h=6
w=8
print('A fruit with height {} and width {} is predicted to be: {}'.
     format(h, w, ['not an apple', 'an apple'][clf.predict([[h,w]])[0]]))
h=10
w=7
print('A fruit with height {} and width {} is predicted to be: {}'.
     format(h, w, ['not an apple', 'an apple'][clf.predict([[h,w]])[0]]))
subaxes.set_xlabel('height')
subaxes.set_ylabel('width')
print('Accuracy on training set: {}'.format(clf.score(X_train, y_train)))
print('Accuracy on test set: {}'.format(clf.score(X_test, y_test)))
```
